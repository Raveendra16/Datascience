```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('final.csv')

# Specify player name
player_name = 'Australia'

# Filter data for the specified player
player_df = df[df['opponent'] == player_name]
player_df=df.groupby(df['opponent'])['index'].count().reset_index()

# Plot the scores
# player_df=player_df.iloc[1:15]
plt.plot(player_df['index'], player_df['runs'], marker='o')
plt.xlabel('Index')
plt.ylabel('Scores')
plt.title('Scores for {}'.format(player_name))
plt.show()
```
